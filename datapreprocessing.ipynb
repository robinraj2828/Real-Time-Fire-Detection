```python
import numpy as np
import cv2
import os

os.mkdir('./FD_data/')
os.mkdir('./FD_data/fire/')
os.mkdir('./FD_data/nonfire/')
count = 1
for img in os.listdir('/kaggle/input/fire-detection/fire/fire/'):
    image = cv2.imread('/kaggle/input/fire-detection/fire/fire/'+img)
    cv2.imwrite('./FD_data/fire/'+str(count)+'.jpg',image)
    count+=1
count = 1
for img in os.listdir('/kaggle/input/fire-detection/non-fire/nonfire/nonfire/'):
    image = cv2.imread('/kaggle/input/fire-detection/non-fire/nonfire/nonfire/'+img)
    cv2.imwrite('./FD_data/nonfire/'+str(count)+'.jpg',image)
    count+=1    
```
